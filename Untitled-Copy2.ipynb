```python
import scvelo as scv
import scanpy as sc
import numpy as np 
```

```python
scv.settings.verbosity = 3  
scv.settings.presenter_view = True 
scv.set_figure_params('scvelo') 
```

```python
adata1 = scv.datasets.pancreas()
```
